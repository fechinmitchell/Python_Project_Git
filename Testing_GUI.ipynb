```python
#Written by Fechin Mitchell on 25/03/2019 Version 1.0
import os
import tkinter as tk
from tkinter import filedialog
from tkinter import *
from PIL import ImageTk, Image
import subprocess as sub

directory = ('C:\\Users\\fmitchell\\Python_Version_Project\\XML_Files_T02_T05\\T02\\GENISYS_T02_1')
WANTED = 5

sucMes = 'Run Successful'

def UploadAction(event=None):
    filename = filedialog.askopenfilename()
    print('Selected:', filename)
    
    for filename in os.listdir(directory):
        if filename.endswith(".xml"):
            print(os.path.join(directory, filename))
            with open("C:\\Users\\fmitchell\\Python_Version_Project\\XML_Files_T02_T05\\T02\\GENISYS_T02_1\\" + filename) as XMLFile:  
                for line in XMLFile:
                    left,sep,right = line.partition('configuration_version')
                    if  sep: # True iff 'Figure' in line
                        print(right[:WANTED])
                    
                    
                        with open( "New_File","a") as myfile:
                            myfile.write(filename + ", " + (right[:WANTED]) + "\n") 
                            print(sucMes)                          
                            
            continue
        else:
            continue
```
